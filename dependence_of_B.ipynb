# Plot of dependence of B

```python
# Only consider 50000 points 

A = B_df[(B_df['N']==50000) & 
         (B_df['model'].isin(models))]

# make an explicit copy as we will be changing things
B = A.copy()

# create label
CC = {'RIPS1':0,'RIPS2':1,'ALPHA1':2,'ALPHA2':3,'ALPHA3':4,'ALPHA4':5}
CCR = {0:'${\cal T}$ = Rips / k=1',\
       1:'${\cal T}$ = Rips / k=2',\
       2:'${\cal T}$ = Čech / k=1',\
       3:'${\cal T}$ = Čech / k=2', \
       4:'${\cal T}$ = Čech / k=3',\
       5:'${\cal T}$ = Čech / k=4'}

# compute color based on d,k and type
B['color']=B.apply(lambda row : CC[row['type']+str(row['homdim'])], axis = 1)

# average
mu=A.groupby(['homdim','type','dim',])['B'].mean()
mu=mu.reset_index()

# variance
vr=A.groupby(['homdim','type','dim',])['B'].std()
vr=vr.reset_index()

# compute colors
mu['color']=mu.apply(lambda row : CC[row['type']+str(row['homdim'])], axis = 1)
vr['color']=vr.apply(lambda row : CC[row['type']+str(row['homdim'])], axis = 1)
```

```python
# The same plot for stratified sampling
StratSD = pd.read_pickle("./summary_files/Strat.pkl")
StratDF = StratSD[(StratSD['homdim']==1) &\
          (StratSD['type']=='RIPS') & 
          (StratSD['model']=='stratifiedDD')]

# compute effective dimension depending on the ratio - extracted 
# from the file name.
r= np.array([float(row['dirname'].split('/')[5]) for i,row in StratDF.iterrows()])
r[r<10] = r[r<10]/10
r[r>10] = 1/r[r>10]
# add effective dimension
StratDF_full=pd.DataFrame(StratDF)
StratDF_full['effdim'] =2+(1-r)
    
```

```python
# plot full all - with stratified
# -------------------------------

plt.figure(figsize=(10,8))
ax = plt.subplot(1,1,1)

# first just points - stratified points
for i in range(5):
    idx = mu['color']==i
    base_line=ax.plot(mu.loc[idx]['dim'],\
                      mu.loc[idx]['B'],\
                      'o',\
                      markersize=10,\
                      label=CCR[i])
    idy = vr['color']==i
    ax.errorbar(mu.loc[idx]['dim'],\
                mu.loc[idx]['B'],\
                yerr=vr.loc[idy]['B'],\
                capsize=20,\
                linewidth=4,\
                c=base_line[0].get_color())

ax.plot(StratDF_full['effdim'],StratDF_full['B'],'o',markersize=10,label='Stratified \n ${\cal T}$ = Rips / k=1')
FN_TITLE = 30
FN_AXES = 24
FN_LEGEND = 24
ax.tick_params(axis='x', labelsize=16)
ax.tick_params(axis='y', labelsize=16)

# plot the remaining
ax.set_xlabel('d', fontsize=24)
ax.set_ylabel('B', fontsize=24)
leg=ax.legend(loc='upper left', fontsize=16)
for lh in leg.legendHandles: 
    lh.set_markersize(12)
plt.tight_layout()
plt.savefig('B_all_strat.pdf', dpi=300,bbox_inches = "tight",pad_inches=0.1)
```
